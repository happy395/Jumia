## Webscraping jumia.com.ng/laptops site. This is to get information about the different models of laptops available on the site and their prices.

```python
#import the necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#website to get the data from
website = 'https://www.jumia.com.ng/laptops/'
```

```python
#send a 'GET' request to the website
response = requests.get(website)
```

```python
#code to know if website is scrapable
response.status_code
```

### if it returns 200, then the site allows access to get data from it.

```python
#create a parse tree to extract data from HTML
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
#To get information on the 1st laptop in the site
laptop_name = soup.find(class_ = 'name').get_text()
laptop_name
```

```python
#All laptops in the 1st page of the site
all_laptop_name = soup.findAll(class_ = 'name')
all_laptop_name
```

```python
#Loop through the 1st page to get the laptops on that page
lap_name = []

for i in all_laptop_name:
    lap_name.append(i.get_text())
    
lap_name
```

```python
#info about the 1st laptop price
laptop_price = soup.find(class_ = 'prc').get_text()
laptop_price
```

```python
#info about all laptop prices in 1st page of the website
all_laptop_prices = soup.findAll(class_ = 'prc')
all_laptop_prices
```

```python
#loop through the 1st page to get all laptop prices
lap_price = []

for p in all_laptop_prices:
    lap_price.append(p.get_text())
    
lap_price
```

```python
#put the data in a pandas DataFrame
laptop_info = {'Laptop Name': lap_name, 'Laptop Price':  lap_price}
df = pd.DataFrame(laptop_info)
df
```

```python
#drop rows with no values
df.drop(index = [40, 41, 42, 43 , 44, 45, 46, 47], inplace=True)
```

```python
#test if the rows with no data have been removed
df
```

### Scraping all 50 pages

```python
#Get the laptop names and prices for all laptops in the site and put the data in a DataFrame
laptop_info = pd.DataFrame(columns=['Laptop Name', 'Laptop Price'])

for i in range(1, 51):
    url = 'https://www.jumia.com.ng/laptops/?page='+str(i)+'#catalog-listing'
    response = requests.get(url)

    soup = BeautifulSoup(response.content, 'html.parser')
    
    laptop_name = soup.findAll(class_ = 'name')
    all_laptop_names = []
    
    for n in laptop_name:
        all_laptop_names.append(n.get_text())
        
    laptop_price = soup.findAll(class_ = 'prc')
    all_laptop_prices = []
    
    for p in laptop_price:
        all_laptop_prices.append(p.get_text())
    
    for k in range(len(all_laptop_names)):
        laptop_info = laptop_info.append({'Laptop Name': all_laptop_names[k], 'Laptop Price': all_laptop_prices[k]},
                                         ignore_index=True)
```

```python
laptop_info
```

```python
#drop rows with no values
laptop_info.drop(index=[40, 41, 42, 43, 44, 45, 46, 47], inplace=True)
```
